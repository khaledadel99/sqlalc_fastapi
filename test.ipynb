```python
import pandas as pd

df_customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'signup_date': pd.to_datetime(['2023-01-01', '2023-01-15', '2023-02-01', '2023-02-20'])
})

df_transactions = pd.DataFrame({
    'transaction_id': [101, 102, 103, 104, 105, 106, 107],
    'customer_id': [1, 1, 2, 2, 3, 3, 4],
    'amount': [200, 300, 150, 100, 400, 100, 500],
    'transaction_date': pd.to_datetime([
        '2023-01-02', '2023-01-05', '2023-01-20', '2023-01-22',
        '2023-02-10', '2023-03-01', '2023-02-25'
    ])
})
```

```python
df = df_customers.merge(df_transactions, on="customer_id", how ="left")
df["diff"] = (df["transaction_date"] - df["signup_date"]).dt.days
df = df[df["diff"].between(0,30)]
df["summ"]=df.groupby("customer_id")["amount"].transform("sum")
df
```
